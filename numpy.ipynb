Frmula para achar os coeficientes angular e linear
-  = coeficiente angular
- n = nmero de elementos
- Y = Moscow
- X = datas



```python
Y = Moscow
X = datas
n = np.size(Moscow)
```

```python
num = (n*np.sum(X*Y) - np.sum(X)*np.sum(Y))
den = (n*np.sum(np.power(X,2)) - np.power(np.sum(X),2))
a = num / den
a
```

Para calcular o coeficiente linear, usamos a seguinte frmula:



```python
b = np.mean(Y) - (a * np.mean(X))
y = a*X + b
```

```python
np.linalg.norm(Moscow-y)
```

```python
plt.plot(datas,Moscow)
plt.plot(x,y)
```

```python
np.random.randint(low=40,high=100,size=100) 
# gera inteiros aleatorios

coef_angulares = np.random.uniform(low=0.10,high=0.90,size=100) 
# gera floats aleatorios
```

```python
norma = np.array([])
for i in range(100):
    b = np.mean(Y) - (coef_angulares[i] * np.mean(X))
    norma = np.append(norma, np.linalg.norm(Moscow-(coef_angulares[i]*X+b)))
    
np.min(norma) # teste de qualidade da reta
```
